```python
import numpy as np
import pandas as pd
import xarray as xr
import getpass

user = getpass.getuser()

from itertools import product
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ce = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damages_ce.csv")
mc = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/damages_mc.csv")
ce = ce[["discrate","year","pulse","rcp","ce_damages", "variable"]]
mc = mc[["discrate","year","pulse","rcp","mc_damages", "variable"]]
```

```python
damages = pd.merge(ce, mc, on= ['discrate', 'year', 'pulse', 'rcp'], how = 'inner')
damages['diff_damages'] = damages.ce_damages - damages.mc_damages
damages
```

```python
damages[(damages['diff_damages'] < 0)].groupby('discrate')['diff_damages'].describe()
```

```python
# generating a list of years where diff damages is negative
years_low = damages.loc[damages.diff_damages < 0].year.unique()
years_low
```

```python
# reading the coeffs in ce and mc to obtain damage functions

ce_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/ce_df_coeffs_SSP3.csv")
ce_coeffs = ce_coeffs[["year", "cons","beta1","beta2"]]

mc_col_sm_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damage_function_comparison/smooth_anomalies_df_collapsed_output_SSP3.csv")
mc_col_sm_coeffs = mc_col_sm_coeffs[["year", "cons","beta1","beta2"]]

mc_sm_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damage_function_comparison/smooth_anomalies_df_mean_output_SSP3.csv")
mc_sm_coeffs = mc_sm_coeffs[["year", "cons","beta1","beta2"]]

mc_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/df_mean_output_SSP3.csv")
mc_coeffs = mc_coeffs[["year", "cons","beta1","beta2"]]
```

```python
# generating df of GMST anomalies between -5 to 20 in 1101 steps for each year between 2015 to 2100
temp_range = np.linspace(-5, 20, 1101)
idx = product(range(2015,2101), temp_range) 
temps = np.stack([temp_range for _ in range(2015,2101)]).flatten() 
temps = pd.DataFrame(temps, index=[i for i,x in idx], columns=['temp'])
temps = temps.reset_index().rename(columns={"index":"year"})
# temps # checking things out
```

```python
# certainty equivalence damage function
ce_df = pd.merge(ce_coeffs, temps, on= ['year'], how = 'inner') 
ce_df['ce_df_T'] = (ce_df.cons + ce_df.temp * ce_df.beta1 + ce_df.temp**2 * ce_df.beta2)
# ce_df # checking things out

# non-CE mean labor mc damage function with smoothed anomnalies
mc_col_sm_df = pd.merge(mc_col_sm_coeffs, temps, on= ['year'], how = 'inner') 
mc_col_sm_df['mc_col_sm_df_T'] = (mc_col_sm_df.cons + mc_col_sm_df.temp * mc_col_sm_df.beta1 + mc_col_sm_df.temp**2 * mc_col_sm_df.beta2)
# mc_col_sm_df # checking things out

# non-CE cloud labor mc damage function with smoothed anomnalies
mc_sm_df = pd.merge(mc_sm_coeffs, temps, on= ['year'], how = 'inner')
mc_sm_df['mc_sm_df_T'] = (mc_sm_df.cons + mc_sm_df.temp * mc_sm_df.beta1 + mc_sm_df.temp**2 * mc_sm_df.beta2)
# mc_sm_df # checking things out

# non-CE cloud labor mc damage function
mc_df = pd.merge(mc_coeffs, temps, on= ['year'], how = 'inner')
mc_df['mc_df_T'] = (mc_df.cons + mc_df.temp * mc_df.beta1 + mc_df.temp**2 * mc_df.beta2)
# mc_sm_df # checking things out
```

```python
# overlaid grid plot of all years where ce_df_T < mc_df_T as per the SCC calculation

# comparison of smooth anomalies CE, non-CE cloud, and non-CE mean 
df1 = pd.merge(pd.merge(ce_df, mc_sm_df, on = ['year', 'temp'], how = 'outer'), mc_col_sm_df, on = ['year', 'temp'], how = 'outer')
df1 = df1[df1['year'].isin(years_low)]

years = list(range(2020, 2101, 10))
# comparison of non-CE cloud smooth vs non-smooth anomalies
df2 = pd.merge(mc_sm_df, mc_df, on = ['year', 'temp'], how = 'outer')
df2 = df2[df2['year'].isin(years)]
# df2.groupby('year').temp.unique()
```
