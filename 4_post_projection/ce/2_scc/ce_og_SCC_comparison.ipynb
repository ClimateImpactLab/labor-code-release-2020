```python
ce_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/ce_df_coeffs_SSP3.csv")
ce_coeffs = ce_coeffs[["year", "cons","beta1","beta2"]]
mc_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/df_mean_output_SSP3.csv")
mc_coeffs = mc_coeffs[["year", "cons","beta1","beta2"]]
```

```python
# ce_coeffs
```

```python
temp_range = np.arange(0,11)
idx = product(range(2015,2100), temp_range) 
temps = np.stack([temp_range for _ in range(2015,2100)]).flatten() 
temps = pd.DataFrame(temps, index=[i for i,x in idx], columns=['temp'])
temps = temps.reset_index().rename(columns={"index":"year"})
temps
```

```python
ce_df = pd.merge(ce_coeffs, temps, on= ['year'], how = 'outer')
ce_df['damage_function_T'] = (ce_df.cons + ce_df.temp * ce_df.beta1 + ce_df.temp**2 * ce_df.beta2)
ce_df
```

```python
mc_df = pd.merge(mc_coeffs, temps, on= ['year'], how = 'outer')
mc_df['damage_function_T'] = (mc_df.cons + mc_df.temp * mc_df.beta1 + mc_df.temp**2 * mc_df.beta2)
mc_df
```

```python
sns.lineplot(data=mc_df.loc[mc_df.year == 2097], x='temp', y='damage_function_T', label="original")
sns.lineplot(data=ce_df.loc[ce_df.year == 2097], x='temp', y='damage_function_T', label="CE")
plt.legend()
```
