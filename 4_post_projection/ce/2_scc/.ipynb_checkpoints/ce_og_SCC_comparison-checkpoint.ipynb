```python
import numpy as np
import pandas as pd
import xarray as xr
import getpass

user = getpass.getuser()

from itertools import product
import seaborn as sns
import matplotlib.pyplot as plt

ce = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damages_ce.csv")
mc = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/damages_mc.csv")
```

```python
ce = ce[["discrate","year","pulse","rcp","ce_damages", "variable"]]
mc = mc[["discrate","year","pulse","rcp","mc_damages", "variable"]]
```

```python
df = pd.merge(ce, mc, on= ['discrate', 'year', 'pulse', 'rcp'], how = 'outer')
df['diff_damages'] = df.ce_damages - df.mc_damages
df
```

```python
ce_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/ce_df_coeffs_SSP3.csv")
ce_coeffs = ce_coeffs[["year", "cons","beta1","beta2"]]
mc_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/df_mean_output_SSP3.csv")
mc_coeffs = mc_coeffs[["year", "cons","beta1","beta2"]]
```

```python
# ce_coeffs
```

```python
temp_range = np.linspace(-5, 20, 1101)
idx = product(range(2015,2100), temp_range) 
temps = np.stack([temp_range for _ in range(2015,2100)]).flatten() 
temps = pd.DataFrame(temps, index=[i for i,x in idx], columns=['temp'])
temps = temps.reset_index().rename(columns={"index":"year"})
temps
```

```python
ce_df = pd.merge(ce_coeffs[ce_coeffs['year'].between(2021,2025)], temps, on= ['year'], how = 'inner')
ce_df['damage_function_T'] = (ce_df.cons + ce_df.temp * ce_df.beta1 + ce_df.temp**2 * ce_df.beta2)
ce_df
```

```python
mc_df = pd.merge(mc_coeffs[mc_coeffs['year'].between(2021,2025)], temps, on= ['year'], how = 'inner')
mc_df['damage_function_T'] = (mc_df.cons + mc_df.temp * mc_df.beta1 + mc_df.temp**2 * mc_df.beta2)
mc_df
```

```python
# years = df.loc[df.diff_damages < 0].year.unique()
# print(years[2])
df = pd.merge(ce_df, mc_df, on = ['year', 'temp'], how = 'outer')
df
```

```python
g = sns.FacetGrid(df, col="year", col_wrap=5, sharey=True)
g.map_dataframe(sns.lineplot, x='temp', y='damage_function_T_x', label="original").map_dataframe(sns.lineplot, x='temp', y='damage_function_T_y', label="CE")
# sns.lineplot(data=mc_df.loc[mc_df.year == [year], x='temp', y='damage_function_T', label="original").set_title([year])
# sns.lineplot(data=ce_df.loc[ce_df.year == [year], x='temp', y='damage_function_T', label="CE")
# plt.legend()
```
