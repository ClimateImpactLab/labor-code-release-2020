```python
import numpy as np
import pandas as pd
import xarray as xr
import getpass

user = getpass.getuser()

from itertools import product
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# change file name here for RCP8.5
ce = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damages_ce_45.csv") 
mc = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/damages_mc.csv")
mc_smooth = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damage_function_comparison/smooth_damages_mc_45.csv")
ce = ce[["discrate","year","pulse","rcp","ce_damages", "variable"]]
mc = mc[["discrate","year","pulse","rcp","mc_damages", "variable"]]
mc_smooth = mc_smooth[["discrate","year","pulse","rcp","mc_damages", "variable"]]
```

```python
damages = pd.merge(ce, mc_smooth, on= ['discrate', 'year', 'pulse', 'rcp'], how = 'inner') #change mc to mc_smooth to compare damages from 
damages['diff_damages'] = damages.ce_damages - damages.mc_damages
damages
```

```python
damages.groupby('discrate')['diff_damages'].describe()
```

```python
damages[(damages['diff_damages'] < 0)].groupby('discrate')['diff_damages'].describe()
```

```python
# generating a list of years where diff damages is negative
years_low = damages.loc[damages.diff_damages < 0].year.unique()
years_low
```

```python
# reading the coeffs in ce and mc to obtain damage functions

ce_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/ce_df_coeffs_SSP3.csv")
ce_coeffs = ce_coeffs[["year", "cons","beta1","beta2"]]

mc_col_sm_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damage_function_comparison/smooth_anomalies_df_collapsed_output_SSP3.csv")
mc_col_sm_coeffs = mc_col_sm_coeffs[["year", "cons","beta1","beta2"]]

mc_sm_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/damage_function_comparison/smooth_anomalies_df_mean_output_SSP3.csv")
mc_sm_coeffs = mc_sm_coeffs[["year", "cons","beta1","beta2"]]

mc_coeffs = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/damage_function_mc/df_mean_output_SSP3.csv")
mc_coeffs = mc_coeffs[["year", "cons","beta1","beta2"]]
```

```python
# checking why the difference between ce damages and mc damages is negative:

anom = 0.61996651 # 0.62153167 from pulse

df = ce_coeffs.loc[ce_coeffs['year'] == 2032].copy()
df['var'] = "CE smooth"
df = df.append(mc_coeffs.loc[mc_coeffs['year'] == 2032]) # change mc_coeffs here as per the requirement
df['var'] = df['var'].fillna("non CE non smooth")
df['damages'] = (df.cons + anom * df.beta1 + anom**2 * df.beta2)

df[['var', 'damages', 'cons', 'beta1', 'beta2', 'year']]
```

```python
# ANALYSIS FOR RCP4.5

coeffs = pd.merge(ce_coeffs, mc_sm_coeffs, on= ['year'], how = 'inner') 
# coeffs = coeffs[coeffs['year'].isin(years_low)] # uncomment if comparison wanted only for years where ce damages < regular damages

fair = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/rcp45_anomalies.csv")

df = pd.merge(coeffs, fair, on= ['year'], how = 'inner')

# generate CE variables: damages due to pulse and level at rcp anomaly 
df['pulse_df_ce'] = (df.cons_x + df.fair_temperatures_anomaly_pulse * df.beta1_x + df.fair_temperatures_anomaly_pulse**2 * df.beta2_x)
df['level_ce'] = (df.cons_x + df.fair_temperatures_anomaly_rcp * df.beta1_x + df.fair_temperatures_anomaly_rcp**2 * df.beta2_x)
df['ce_damages'] = df.pulse_df_ce - df.level_ce

# generate non-CE variables: damages due to pulse and level at rcp anomaly 
df['pulse_df_mc'] = (df.cons_y + df.fair_temperatures_anomaly_pulse * df.beta1_y + df.fair_temperatures_anomaly_pulse**2 * df.beta2_y)
df['level_reg'] = (df.cons_y + df.fair_temperatures_anomaly_rcp * df.beta1_y + df.fair_temperatures_anomaly_rcp**2 * df.beta2_y)
df['reg_damages'] = df.pulse_df_mc - df.level_reg

# diff between CE and non-CE damages
df['diff'] = df.ce_damages - df.reg_damages

# slope = beta1 + 2*beta2*fair_anomaly for each case for rcp anomaly
df['slope_ce'] = (df.beta1_x + 2 * df.fair_temperatures_anomaly_rcp * df.beta2_x)
df['slope_reg'] = (df.beta1_y + 2 * df.fair_temperatures_anomaly_rcp * df.beta2_y)

# final df
df[['year', 'ce_damages', 'reg_damages', 'fair_temperatures_anomaly_rcp', 'diff', 'level_ce', 'level_reg', 'slope_ce', 'slope_reg']]#.to_csv('/home/nsharma/repos/labor-code-release-2020/output/ce/damage_function_comparison/rcp45_df.csv')
```

```python
# generating df of anomalies between -5 to 20 in 1101 steps for each year between 2015 to 2100
temp_range = np.linspace(-5, 20, 1101)
idx = product(range(2015,2101), temp_range) 
temps = np.stack([temp_range for _ in range(2015,2101)]).flatten() 
temps = pd.DataFrame(temps, index=[i for i,x in idx], columns=['temp'])
temps = temps.reset_index().rename(columns={"index":"year"})
# temps # checking things out
```

```python
# certainty equivalence damage function
ce_df = pd.merge(ce_coeffs, temps, on= ['year'], how = 'inner') 
ce_df['ce_df_T'] = (ce_df.cons + ce_df.temp * ce_df.beta1 + ce_df.temp**2 * ce_df.beta2)
# ce_df # checking things out

# non-CE cloud labor mc damage function
mc_df = pd.merge(mc_coeffs, temps, on= ['year'], how = 'inner')
mc_df['mc_df_T'] = (mc_df.cons + mc_df.temp * mc_df.beta1 + mc_df.temp**2 * mc_df.beta2)
# mc_sm_df # checking things out

# non-CE mean labor mc damage function with smoothed anomnalies
mc_col_sm_df = pd.merge(mc_col_sm_coeffs, temps, on= ['year'], how = 'inner') 
mc_col_sm_df['mc_col_sm_df_T'] = (mc_col_sm_df.cons + mc_col_sm_df.temp * mc_col_sm_df.beta1 + mc_col_sm_df.temp**2 * mc_col_sm_df.beta2)
# mc_col_sm_df # checking things out

# non-CE cloud labor mc damage function with smoothed anomnalies
mc_sm_df = pd.merge(mc_sm_coeffs, temps, on= ['year'], how = 'inner')
mc_sm_df['mc_sm_df_T'] = (mc_sm_df.cons + mc_sm_df.temp * mc_sm_df.beta1 + mc_sm_df.temp**2 * mc_sm_df.beta2)
# mc_sm_df # checking things out
```

```python
# overlaid grid plot of all years where ce_df_T < mc_df_T as per the SCC calculation

# comparison of smooth anomalies: CE, non-CE cloud, and non-CE mean 
df1 = pd.merge(pd.merge(ce_df, mc_sm_df, on = ['year', 'temp'], how = 'outer'), mc_col_sm_df, on = ['year', 'temp'], how = 'outer')
df1 = df1[df1['year'].isin(years_low)]

# generating years from 2020 to 2100 at every ten years
years = list(range(2020, 2301, 10))

# comparison of non-CE cloud smooth vs non-smooth anomalies
df2 = pd.merge(ce_df, mc_sm_df, on = ['year', 'temp'], how = 'outer')
df2 = df2[df2['year'].isin(years)]

# comapre CE vs non-CE cloud non-smooth DFs
df3 = pd.merge(ce_df, mc_df, on = ['year', 'temp'], how = 'outer')
df3 = df3[df3['year'].isin(years_low)]

fair_anom = pd.read_csv("/home/nsharma/repos/labor-code-release-2020/output/ce/rcp85_anomalies.csv")
df4 = pd.merge(df3, fair_anom, on = ['year'], how = 'outer') 
```

```python
# grid plot of non-CE cloud non-smooth vs CE
g = sns.FacetGrid(df4, col="year", col_wrap=6, sharey=True)
g.map_dataframe(sns.lineplot, x='temp', y='ce_df_T', color = "blue", label="CE smooth anomalies").map_dataframe(sns.lineplot, x='temp', y='mc_df_T', color = "red", label="non-CE cloud non smooth anomalies")
lgd = plt.legend(bbox_to_anchor=(-1,9.5), ncol=3)
# plt.savefig('/home/nsharma/repos/labor-code-release-2020/output/ce/damage_func_comparison_vline.pdf', format='pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
```
