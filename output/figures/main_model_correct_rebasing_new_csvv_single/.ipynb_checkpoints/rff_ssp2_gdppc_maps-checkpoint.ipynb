```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import geopandas as gpd
import pandas as pd
import xarray as xr
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
data_dir = '/shares/gcp/social/rff/'

data = pd.read_csv(f'{data_dir}/gdppc-nohier-6546.csv', skiprows = 10)
df_plot = data.rename(columns={'iso': 'ISO', 'value': 'gdppc'}).loc[(data['year'] == 2100)]
# df_plot.to_csv(f'{data_dir}/df_plot.csv')
```

```python
shp_file = gpd.read_file(
    '/shares/gcp/estimation/labor/code_release_int_data/shapefiles/world-combo-new-nytimes/new_shapefile.shp')

shp_file = shp_file.set_index('hierid')
shp_file.index.name = 'region' 
```

```python
def make_map(df,
             colname,
             title,
             name_file,
             normalize=False,
             step_cb=15000,
             round_param_left=-2,
             round_param_right=None,
             figsize=(30, 15),
             color_scale='YlOrRd',
             color_max=None,
             save_path='/shares/gcp/social/rff'):
    
    # create figure
    fig, ax = plt.subplots(figsize=figsize, facecolor='white')
    
    color_max = df[colname].max() if color_max == None else color_max
    
    # resize colourbar
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="2%", pad=0.2)
    ax.figure.axes[1].tick_params(labelsize=10)
        
    # Plotting function -- from geopandas
    ax = df.plot(column=colname,
                 cmap=color_scale,
                 legend=True,
                 norm=mpl.colors.Normalize(vmin=0, vmax=color_max),
                 ax=ax,
                 cax=cax)

    ax.set_title(title, fontsize=10)
    ax.set_axis_off()
    plt.show()
    fig.savefig(f'{save_path}/{name_file}', dpi = 300, bbox_inches = 'tight', pad_inches = 0)
    
```

```python
year = 2100

make_map(df=shp_file.merge(df_plot, on='ISO'),
     colname='gdppc',
     title= 'RFF GDPpc in 2100, 2019 USD',
     name_file=f'gdppc_RFF_{year}.png',
     color_max = 400000,
     figsize=(20, 30),
     color_scale='Blues')
```
